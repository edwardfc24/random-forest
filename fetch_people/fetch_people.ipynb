# nearest neighbords


```python
from sklearn import neighbors
from PIL import Image
from sklearn.model_selection import train_test_split
from scipy import misc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Cargo el data frame

```python
from sklearn.datasets import fetch_lfw_people
#df = pd.DataFrame(fetch_lfw_people())
fetch_people = fetch_lfw_people()
x_data = fetch_people.images
y_data = fetch_people.target
```

Cargar imagen

```python
def load_image( infilename ) :
    img = Image.open( infilename )
    img =img.convert('L')
    img.load()
    data = np.asarray( img, dtype="int32" )
    return data
```

```python
plt.imshow(load_image('imagenes prueba\Sergei_Alexandrovitch_Ordzhonikidze_0001.jpg'), cmap='Greys_r')
plt.show()
```

# Entrenar

```python
k_nearest_classifier = neighbors.KNeighborsClassifier(15)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.20)
```

```python
k_nearest_classifier.fit(x_train, y_train)
```

```python
results = k_nearest_classifier.score(x_train, y_train)
print(f"Grado de Confianza: {results}")
```

Grado de Confianza: 0.18940109578688835
