```python
import pandas as pd
from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
people = fetch_lfw_people()
```

```python
k_nearest = neighbors.KNeighborsClassifier(n_neighbors=10)
```

```python
def show_image(index):
    plt.imshow(people['images'][index], cmap='Greys_r')
    plt.show()

def test_case(index):
    show_image(index)
    test = np.array(people['data'][index]).reshape(1,-1)
    number_predicted = k_nearest.predict(test)
    print(people['target_names'][number_predicted[0]])
```

```python
data_x = people.data
data_y = people.target
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.2)
```

```python
k_nearest.fit(test_x, test_y)
```

```python
test_case(666) # 666, 90, 55, 665
```

```python
score = k_nearest.score(test_x,test_y)
print(f"The score is: {score}")
```
