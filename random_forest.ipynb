```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
digits = datasets.load_digits()
df = pd.DataFrame(digits.data)
train_x = df
train_y = digits.target
```

```python
r_forest_classifier = RandomForestClassifier()
```

```python
r_forest_classifier.fit(train_x, train_y)
```

Funciones para la visualizacion de informacion

```python
def show_image(index):
    plt.imshow(digits.images[index], cmap="Greys_r")
    plt.show()
#def show_image(index, source):
#    plt.imshow(source['images'][index], cmap='Greys_r')
#    plt.show()


def test_case(index):
    show_image(index)
    test = np.array(digits.data[index]).reshape(1,-1)
    number_predict = r_forest_classifier.predict(test)
    print(digits.target_names[number_predict[0]])
```

```python
test_case(1650)
```

```python
score = r_forest_classifier.score(train_x, train_y)
print(f"Score random Forest: {score}")
```
